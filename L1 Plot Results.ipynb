```python
import numpy as np
import matplotlib.pyplot as plt

c0_0001 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c0_001 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c0_01 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c0_1 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c0_5 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c1 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c5 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c10 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c50 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
c100 = {"Training Size (Users)":[], "Training Time (s)":[], "Psuedo Log-likelihood Training": [], "Psuedo Log-likelihood Testing": []}
with open("rawResults.txt", "r") as result:
    for x in result:
        x = x.strip()
        x = x.split()
        if float(x[1]) == 0.0001:
            c0_0001["Training Size (Users)"].append(int(x[0]))
            c0_0001["Training Time (s)"].append(float(x[2]))
            c0_0001["Psuedo Log-likelihood Training"].append(float(x[3]))
            c0_0001["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 0.001:
            c0_001["Training Size (Users)"].append(int(x[0]))
            c0_001["Training Time (s)"].append(float(x[2]))
            c0_001["Psuedo Log-likelihood Training"].append(float(x[3]))
            c0_001["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 0.01:
            c0_01["Training Size (Users)"].append(int(x[0]))
            c0_01["Training Time (s)"].append(float(x[2]))
            c0_01["Psuedo Log-likelihood Training"].append(float(x[3]))
            c0_01["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 0.1:
            c0_1["Training Size (Users)"].append(int(x[0]))
            c0_1["Training Time (s)"].append(float(x[2]))
            c0_1["Psuedo Log-likelihood Training"].append(float(x[3]))
            c0_1["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 0.5:
            c0_5["Training Size (Users)"].append(int(x[0]))
            c0_5["Training Time (s)"].append(float(x[2]))
            c0_5["Psuedo Log-likelihood Training"].append(float(x[3]))
            c0_5["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 1:
            c1["Training Size (Users)"].append(int(x[0]))
            c1["Training Time (s)"].append(float(x[2]))
            c1["Psuedo Log-likelihood Training"].append(float(x[3]))
            c1["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 5:
            c5["Training Size (Users)"].append(int(x[0]))
            c5["Training Time (s)"].append(float(x[2]))
            c5["Psuedo Log-likelihood Training"].append(float(x[3]))
            c5["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 10:
            c10["Training Size (Users)"].append(int(x[0]))
            c10["Training Time (s)"].append(float(x[2]))
            c10["Psuedo Log-likelihood Training"].append(float(x[3]))
            c10["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 50:
            c50["Training Size (Users)"].append(int(x[0]))
            c50["Training Time (s)"].append(float(x[2]))
            c50["Psuedo Log-likelihood Training"].append(float(x[3]))
            c50["Psuedo Log-likelihood Testing"].append(float(x[4]))
        elif float(x[1]) == 100:
            c100["Training Size (Users)"].append(int(x[0]))
            c100["Training Time (s)"].append(float(x[2]))
            c100["Psuedo Log-likelihood Training"].append(float(x[3]))
            c100["Psuedo Log-likelihood Testing"].append(float(x[4]))
```

```python
plt.figure(figsize=(8,5));
plt.plot(c0_0001["Training Size (Users)"], c0_0001["Training Time (s)"], color='r', label='C = 0.0001')
plt.plot(c0_001["Training Size (Users)"], c0_001["Training Time (s)"], color='g', label='C = 0.001')
plt.plot(c0_01["Training Size (Users)"], c0_01["Training Time (s)"], color='y', label='C = 0.01')
plt.plot(c0_1["Training Size (Users)"], c0_1["Training Time (s)"], color='b', label='C = 0.1')
plt.plot(c1["Training Size (Users)"], c1["Training Time (s)"], color='m', label='C = 1')
plt.plot(c10["Training Size (Users)"], c10["Training Time (s)"], color='gray', label='C = 10')
plt.title("L1 Training Time vs Training Size by C (Sparsity Penalty)")
plt.xlabel("Training Size (Users)")
plt.ylabel("Training Time (s)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5));
plt.plot(c0_0001["Training Size (Users)"], c0_0001["Psuedo Log-likelihood Testing"], color='r', label='C = 0.0001')
plt.plot(c0_001["Training Size (Users)"], c0_001["Psuedo Log-likelihood Testing"], color='g', label='C = 0.001')
plt.plot(c0_01["Training Size (Users)"], c0_01["Psuedo Log-likelihood Testing"], color='y', label='C = 0.01')
plt.plot(c0_1["Training Size (Users)"], c0_1["Psuedo Log-likelihood Testing"], color='b', label='C = 0.1')
plt.plot(c1["Training Size (Users)"], c1["Psuedo Log-likelihood Testing"], color='m', label='C = 1')
plt.plot(c10["Training Size (Users)"], c10["Psuedo Log-likelihood Testing"], color='gray', label='C = 10')
plt.title("L1 Penalized: Psuedo Log-likelihood Testing vs Training Size by C (Sparsity Penalty)")
plt.xlabel("Training Size (Users)")
plt.ylabel("Psuedo Log-likelihood Training")
plt.grid(True)
plt.legend()
plt.show()
```
